```python
import pandas as p
import matplotlib.pyplot as pl

data = p.read_csv('crypto_data.csv', header=[0,1], index_col=0)
data.head()
```

```python
data.info()
data.describe()
```

```python
pl.figure(figsize=(10,5))
for coin in data.columns.levels[0]:
  pl.plot(data.index, data[coin]['Close'], label=coin)
pl.title('Crypto closin prices 2023-2025')
pl.xlabel('date')
pl.ylabel('price (usd)')
pl.legend()
pl.show()
```

```python
close_prices = p.DataFrame({coin: data[coin]['Close'] for coin in data.columns.levels[0]})
corr = close_prices.corr()
print(corr)
```

```python
import seaborn as s
s.heatmap(corr, annot=True, cmap='coolwarm')
pl.title("crypto price correlation")
pl.show()
```
