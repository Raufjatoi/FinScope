```python
import pandas as p

data = p.read_csv('crypto_scaled.csv', header=[0,1], index_col=0)

data
```

```python
coin = 'BTC-USD'
df = data[[coin]].copy()
```

```python
df['Prev_1'] = df[coin].shift(1)
df['Prev_2'] = df[coin].shift(2)
df['Prev_3'] = df[coin].shift(3)
```

```python
df['MA_5'] = df[coin].rolling(5).mean()
df['MA_10'] = df[coin].rolling(10).mean()
```

```python
df = df.dropna()
```

```python
X = df.drop(coin, axis=1)
y = df[coin]
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, shuffle=False)
```
